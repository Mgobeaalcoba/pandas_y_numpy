# Agregar y eliminar datos en Pandas

```python
df_books.head(2)
```

```python
# Eliminar columnas (axis 1)
df_books.drop('Genre', axis=1).head(2) # De esta forma solo lo voy a borrar de la salida que haré debajo. 
```

```python
df_books.drop('Genre', axis=1, inplace=True) # Con inplace igual True lo borro de manera definitiva en el objeto dataframe
```

```python
df_books.head(2) # Verifico que quedó borrado 
```

```python
# Otra forma de borrar una columna de forma definitiva es guardando nuestro borrado en nuestra variable de dataframe
df_books = df_books.drop('Year', axis=1)
df_books
```

```python
# otra forma de borrar es usando el del de python directamente de la siguiente forma / No es aconsejable igualmente
del df_books['Price']
```
